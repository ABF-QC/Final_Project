# Selecting the stations within a specific Administrative region
---

In this notebook,

### Importing necessary libraries

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd
```

### Define input files

```python
admin_region_file = '../data/Admin_Regions/SHP/regio_s.shp'

# First 3 lines of the csv files must be discarded
rows2skip = list(range(0,3))

stn_inventory = '../data/ECCC/Station_Inventory_EN.csv'

fileout = '../data/ECCC/Station_Inventory_EN_AdminRegion.csv'
```

### Reading the shapefile

```python
admin_regions = gpd.read_file(admin_region_file)
```

```python
admin_regions.plot()
```

```python
admin_regions
```

```python
admin_regions[admin_regions['RES_NM_REG'] == 'Montréal'].plot() 
```

```python
# Verify the projection
admin_regions.crs
```

### Read station inventory file

```python
stns = pd.read_csv(stn_inventory, skiprows=rows2skip)

stns
```

### Keeping only stations in Quebec

```python
stns = stns[stns['Province'] == 'QUEBEC']
stns
```

### Creating a georeference point for each station
CRS = EPSG: 4269 from ../data/Admin_Regions/SHP/regio_s.prj file

```python
stns = gpd.GeoDataFrame(stns, geometry=gpd.points_from_xy(stns['Longitude (Decimal Degrees)'], stns['Latitude (Decimal Degrees)']), crs=4269)
stns
```

### Finding the corresponding Administrative Region for each station

```python
tmp = gpd.sjoin(stns, admin_regions, how="left", predicate="within")
tmp.head()
```

```python
tmp.columns
```

### Adding the corresponding administrative region for each station

```python
stns['admin'] = tmp['RES_NM_REG']
stns
```

### Plot the regions and the stations

```python
# Create figure and adjust its size
fig, ax = plt.subplots(figsize=(14, 14))

# Add admin region to plot
admin_regions.plot(ax=ax, color='lightgray', edgecolor='black', alpha=0.5)

# Plot climate stations
stns.plot(ax=ax,column='admin', cmap='tab20', legend=True, markersize=25, alpha=0.8)

# Add title
ax.set_title('Climate Stations and Administrative Regions of Québec')

# Adjust axis labels
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

# Show plot
plt.show()
```

### Plot the stations over Montreal

```python
# Create figure and adjust its size
fig, ax = plt.subplots(figsize=(14, 14))

# Add admin region to plot
admin_regions[admin_regions['RES_NM_REG'] == 'Montréal'].plot(ax=ax, color='lightgray', edgecolor='black', alpha=0.5)

# Plot climate stations
stns[stns['admin'] == 'Montréal'].plot(ax=ax, legend=False, markersize=25, alpha=0.8)

# Add title
ax.set_title('Climate Stations and Administrative Regions of Québec')

# Adjust axis labels
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

# Show plot
plt.show()
```

### Save data to file

```python
stns.to_csv(fileout, index=False)
```
